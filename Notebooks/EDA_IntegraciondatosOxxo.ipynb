```python
# prompt: me refiero a que crees una columna llamada index año anterior, en el cual hagas la proporcion mes 2024 / mes 2023, y solo rellenes los meses 2024

import pandas as pd
df_ventas['AÑO'] = df_ventas['MES_ID_DATETIME'].dt.year
df_ventas['MES'] = df_ventas['MES_ID_DATETIME'].dt.month

df_ventas_2023 = df_ventas[df_ventas['AÑO'] == 2023].copy()
df_ventas_2024 = df_ventas[df_ventas['AÑO'] == 2024].copy()

df_ventas_2023 = df_ventas_2023.rename(columns={'VENTA_TOTAL': 'VENTA_TOTAL_2023'})
df_ventas_2024 = df_ventas_2024.rename(columns={'VENTA_TOTAL': 'VENTA_TOTAL_2024'})

df_merged = pd.merge(df_ventas_2024, df_ventas_2023[['TIENDA_ID', 'MES', 'VENTA_TOTAL_2023']], on=['TIENDA_ID', 'MES'], how='left')

df_merged['index año anterior'] = df_merged['VENTA_TOTAL_2024'] / df_merged['VENTA_TOTAL_2023']

# Merge the new column back to the original df_ventas dataframe
df_ventas = pd.merge(df_ventas, df_merged[['TIENDA_ID', 'MES_ID', 'index año anterior']], on=['TIENDA_ID', 'MES_ID'], how='left')

df_ventas = df_ventas.drop(columns=['AÑO', 'MES'])
df_ventas
```

```python
# prompt: ayudame generando el promedio y  desviacion estando del index por ID, solo no contemples 12 2024, pero ponle que es INDEX_promedio o desviacion estandar

import pandas as pd
# Excluir el MES_ID 202412 para el cálculo del promedio y desviación estándar del índice
df_ventas_filtered = df_ventas[df_ventas['MES_ID'] != 202412].copy()

# Calcular el promedio del índice por TIENDA_ID, excluyendo 202412
df_ventas_filtered['INDEX_promedio'] = df_ventas_filtered.groupby('TIENDA_ID')['index año anterior'].transform('mean')

# Calcular la desviación estándar del índice por TIENDA_ID, excluyendo 202412
df_ventas_filtered['INDEX_desviacion estandar'] = df_ventas_filtered.groupby('TIENDA_ID')['index año anterior'].transform('std')

# Fusionar las columnas calculadas de vuelta al DataFrame original df_ventas
# Usamos 'left' merge para asegurarnos de que todas las filas de df_ventas se mantengan
df_ventas = pd.merge(df_ventas, df_ventas_filtered[['TIENDA_ID', 'INDEX_promedio', 'INDEX_desviacion estandar']].drop_duplicates(subset=['TIENDA_ID']), on='TIENDA_ID', how='left')

df_ventas
```

## **Prueba Normalidad Index**

La prueba de Lilliefors es una adaptación de la prueba de Kolmogorov-Smirnov diseñada específicamente para casos en los que no conoces los parámetros de la distribución teórica (la media y desviación estándar de la distribución normal ideal) y los estimas a partir de tus propios datos. En tu caso, aunque calculaste la media y desviación estándar de tu muestra, no sabes la media y desviación estándar de la distribución normal verdadera de donde provendrían idealmente tus índices. La prueba de Lilliefors maneja esta situación automáticamente.

```python
# prompt: hazme una funcion que detecte cuales meses de 2023 01 al 2024 12 no estan en nuestro df por ID, y esos inputalos dentro de nuestro df en filas y rellena en base de las columnas estilo fecha correspondientes le reesto dejalo en nans

import pandas as pd
def completar_meses_faltantes(df):
    """
    Detecta los meses entre 202301 y 202412 que faltan para cada TIENDA_ID
    en el DataFrame, los inserta y rellena columnas de fecha.

    Args:
        df (pd.DataFrame): DataFrame de ventas con columnas 'TIENDA_ID' y 'MES_ID'.

    Returns:
        pd.DataFrame: DataFrame con los meses faltantes insertados y
                      columnas de fecha rellenadas, otras columnas en NaN.
    """
    # Asegurarse de que MES_ID sea de tipo cadena para manejarlo como '%Y%m'
    df['MES_ID'] = df['MES_ID'].astype(str)

    # Crear un rango completo de meses de 202301 a 202412
    fechas_completas = pd.date_range(start='2023-01-01', end='2024-12-01', freq='MS').strftime('%Y%m')

    # Obtener todos los TIENDA_ID únicos
    tiendas_unicas = df['TIENDA_ID'].unique()

    # Crear un DataFrame de referencia con todas las combinaciones de TIENDA_ID y MES_ID
    from itertools import product
    referencia_df = pd.DataFrame(list(product(tiendas_unicas, fechas_completas)), columns=['TIENDA_ID', 'MES_ID'])

    # Combinar el DataFrame de referencia con el DataFrame original
    # Esto introduce filas para los meses faltantes, con NaN en las columnas de datos
    df_completo = pd.merge(referencia_df, df, on=['TIENDA_ID', 'MES_ID'], how='left')

    # Rellenar las columnas relacionadas con la fecha para las filas nuevas
    df_completo['MES_ID_DATETIME'] = pd.to_datetime(df_completo['MES_ID'], format='%Y%m', errors='coerce')
    # Handle potential conversion errors if MES_ID is not valid
    df_completo = df_completo.dropna(subset=['MES_ID_DATETIME'])
    df_completo['MES_ID_TIMESTAMP'] = df_completo['MES_ID_DATETIME'].astype('int64') // 10**9

    # Ordenar el DataFrame
    df_completo = df_completo.sort_values(by=['TIENDA_ID', 'MES_ID']).reset_index(drop=True)

    # Para otras columnas como 'PROMEDIO_VENTA', 'DESVIACION_ESTANDAR_VENTA',
    # 'INDEX_promedio', 'INDEX_desviacion estandar' que son por TIENDA_ID
    # y deberían ser consistentes para todos los meses de esa tienda (basado
    # en cómo se calcularon previamente), podemos usar forward fill (fillna(method='ffill'))
    # y backward fill (fillna(method='bfill')) después de ordenar por TIENDA_ID y MES_ID.
    # Esto asume que estos valores no cambian para una TIENDA_ID dada entre meses.
    # Las columnas 'VENTA_TOTAL', 'VENTA_TOTAL_2023', 'VENTA_TOTAL_2024', 'index año anterior'
    # deben quedar como NaN si el mes no existía originalmente.

    # Lista de columnas que queremos propagar (si aplica a tu lógica de negocio)
    cols_to_propagate = [
        'PROMEDIO_VENTA',
        'DESVIACION_ESTANDAR_VENTA',
        'INDEX_promedio',
        'INDEX_desviacion estandar'
    ]

    for col in cols_to_propagate:
        if col in df_completo.columns:
             df_completo[col] = df_completo.groupby('TIENDA_ID')[col].fillna(method='ffill')
             df_completo[col] = df_completo.groupby('TIENDA_ID')[col].fillna(method='bfill')


    return df_completo

# Aplicar la función al DataFrame df_ventas
df_ventas_completo = completar_meses_faltantes(df_ventas.copy())

# Mostrar el resultado
df_ventas_completo
```

```python
# prompt: rellena los datos nans en venta, en base del uin numero random entre promedio venta y contempla la desviacion estandar +-para el rango menor a mayor

import pandas as pd
import numpy as np

def fill_missing_venta_with_random(df):
    """
    Rellena los valores NaN en la columna 'VENTA_TOTAL' de un DataFrame
    con un número aleatorio dentro del rango definido por el promedio
    y la desviación estándar de 'VENTA_TOTAL' para cada 'TIENDA_ID'.

    Args:
        df (pd.DataFrame): DataFrame con columnas 'TIENDA_ID', 'VENTA_TOTAL',
                           'PROMEDIO_VENTA', y 'DESVIACION_ESTANDAR_VENTA'.

    Returns:
        pd.DataFrame: DataFrame con los valores NaN en 'VENTA_TOTAL' rellenados.
    """
    # Identify rows where VENTA_TOTAL is NaN
    nan_mask = df['VENTA_TOTAL'].isna()

    # For rows with NaN in VENTA_TOTAL, calculate the random value
    # based on PROMEDIO_VENTA and DESVIACION_ESTANDAR_VENTA
    # Ensure we only operate on rows with necessary information
    valid_rows_for_fill = df.loc[nan_mask & df['PROMEDIO_VENTA'].notna() & df['DESVIACION_ESTANDAR_VENTA'].notna()].copy()

    if not valid_rows_for_fill.empty:
        # Calculate the lower and upper bounds for the random number
        lower_bound = valid_rows_for_fill['PROMEDIO_VENTA'] - valid_rows_for_fill['DESVIACION_ESTANDAR_VENTA']
        upper_bound = valid_rows_for_fill['PROMEDIO_VENTA'] + valid_rows_for_fill['DESVIACION_ESTANDAR_VENTA']

        # Ensure lower bound is not negative (as sales should be non-negative)
        lower_bound = lower_bound.clip(lower=0)

        # Generate random numbers within the calculated range for each row
        # Use np.random.uniform for floating point random numbers within a range
        random_venta_values = np.random.uniform(low=lower_bound, high=upper_bound, size=len(valid_rows_for_fill))

        # Assign the generated random values back to the DataFrame
        # Use the index of the valid_rows_for_fill to assign correctly
        df.loc[valid_rows_for_fill.index, 'VENTA_TOTAL'] = random_venta_values

    return df

# Apply the function to the DataFrame
df_ventas_completo = fill_missing_venta_with_random(df_ventas_completo)

df_ventas_completo
```
